# Series

```python
list_s = [1,2,-3,6.2, ' data values']
print(list_s)
```

```python
series1 = pd.Series(list_s) #Series create always remember 'S' capital
print(series1)
```

```python
type(series1)
```

```python
series2 = pd.Series([1,2,3,4])
print(series2)
```

```python
empty_s = pd.Series([])
print(empty_s)
```

```python
series3 = pd.Series([1,2,3,4], index = ['a','b','c','d']) #index change
print(series3)
```

```python
series3 = pd.Series([1,2,3,4], index = ['a','b','c','d'], dtype = float) 
#data type change
print(series3)
```

```python
series3 = pd.Series([1,2,3,4], index = ['a','b','c','d'], dtype = float, name = 'data values') 

print(series3)
```

```python
#scalar values means single values
scalar_s = pd.Series(0.5)
print(scalar_s)
```

```python
scalar_s = pd.Series(0.5 , index = [1,2,3])
print(scalar_s)
```

```python
dic_s =pd.Series({'a':1,'b':2})
print(dic_s)
```
